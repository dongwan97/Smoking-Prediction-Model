```python
Columns = [ 'SIDO','SEX', 'AGE_GROUP', 'HEIGHT', 'WEIGHT',
       'WAIST', 'SIGHT_LEFT', 'SIGHT_RIGHT', 'HEAR_LEFT', 'HEAR_RIGHT',
       'BP_HIGH', 'BP_LWST', 'BLDS', 'TOT_CHOLE', 'TRIGLYCERIDE', 'HDL_CHOLE',
       'LDL_CHOLE', 'HMG', 'OLIG_PROTE_CD', 'CREATININE', 'SGOT_AST',
       'SGPT_ALT', 'GAMMA_GTP', 'SMK_STAT_TYPE_CD', 'DRK_YN',
        'CRS_YN', 'TTR_YN']

dataNew=data[Columns]
dataNew
```

```python
dataNew=dataNew.dropna()
```

```python
dataNew=dataNew[data['HEAR_LEFT']!=3]
dataNew=dataNew[data['HEAR_RIGHT']!=3]
dataNew
```

```python
dataNew=dataNew[data['TTR_YN']!=2]
dataNew
```

```python
dataNew['BMI']=dataNew['WEIGHT']/(dataNew['HEIGHT']*dataNew['HEIGHT'])*10000
dataNew
```

```python
dataNew[dataNew['SIGHT_LEFT']==9]=0
dataNew[dataNew['SIGHT_RIGHT']==9]=0
dataNew[['SIGHT_LEFT','SIGHT_RIGHT']]
```

```python
columns_IQR = [  'HEIGHT', 'WEIGHT',
       'WAIST', 'SIGHT_LEFT', 'SIGHT_RIGHT',
       'BP_HIGH', 'BP_LWST', 'BLDS', 'TOT_CHOLE', 'TRIGLYCERIDE', 'HDL_CHOLE',
       'LDL_CHOLE', 'HMG', 'CREATININE', 'SGOT_AST',
       'SGPT_ALT', 'GAMMA_GTP']

Q1 = dataNew.quantile(0.25)
Q1
Q3 = dataNew.quantile(0.75)
IQR = Q3-Q1

dataNew = dataNew[~((dataNew[columns_IQR]<(Q1-3*IQR))|(dataNew[columns_IQR]>(Q3+3*IQR))).any(axis=1)]
dataNew
```
